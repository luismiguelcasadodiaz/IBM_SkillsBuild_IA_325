# Recomendador de plataforma

Una empresa de desarrollo de software desea construir un sistema inteligente que recomiende la plataforma m√°s adecuada (web, mobile o desktop) para nuevos proyectos, en base a las caracter√≠sticas del proyecto y la experiencia previa con proyectos similares.



## ‚úÖ Objetivo

Implementa un sistema en Python utilizando clases, estructuras de datos y machine learning (√°rbol de decisi√≥n) que sea capaz de predecir la mejor plataforma para un nuevo proyecto.



### üß± Parte 1: Modelar los proyectos

  + Crea una clase llamada Project con los siguientes atributos:
    + Atributos
      + name: nombre del proyecto (str)
      + team_size: tama√±o del equipo de desarrollo (int)
      + budget: presupuesto en miles de d√≥lares (float)
      + duration_months: duraci√≥n estimada en meses (int)
      + realtime_required: si requiere tiempo real (bool)
      + needs_offline: si necesita funcionar sin conexi√≥n (bool)
      + target_users: tipo de usuarios objetivo (str: "local", "empresa" o "global")
      + recommended_platform: plataforma recomendada (str: "web", "mobile", "desktop"). Este dato estar√° disponible solo para proyectos hist√≥ricos, no para nuevos.

    + M√©todos:
      + üîß  to_features(self, label_encoder=None) para convertir un proyecto a una **lista** de caracter√≠sticas num√©ricas.



### üìä Parte 2: Construir el dataset

  + Crea una clase llamada ProjectDataset:
    + Atributos:
      + Reciba una lista de proyectos (hist√≥ricos y nuevos).

    + M√©todos:
      + _fit_encoders() para preparar los codificadores.
      + get_X_y() para obtener los datos de entrenamiento.
      + decode_platform(label) para decodificar una predicci√≥n num√©rica.

  Use LabelEncoder para convertir target_users y recommended_platform a variables num√©ricas.

### üß† Parte 3: Entrenar el modelo

  + Crea una clase PlatformRecommender:
    + M√©todos:
      + train(dataset: ProjectDataset): entrena un modelo DecisionTreeClassifier con los datos hist√≥ricos.

      + predict(project: Project): predice la plataforma m√°s adecuada para un nuevo proyecto.
