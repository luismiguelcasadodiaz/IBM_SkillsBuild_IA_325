### Ejercicio 5

#### Ejercicio 5.1
Muestra los primeros registros del siguiente conjunto de datos

```python
df = sns.load_dataset("penguins")
# Resuelve aqui el ejercicio 5.1
df.head(5)
```

#### Ejercicio 5.2
Separa el dataset provisto en conjunto de entrenamiento y de test, dando un 40% de los datos al conjunto de test. La variable objetivo corresponde a la columna "species"

```python
# Resuelve aquí el ejericio 5.2
# Detectamos los nulos


# species	           0
# island	           0
# bill_length_mm	   2
# bill_depth_mm	     2
# flipper_length_mm	 2
# body_mass_g	       2
# sex	              11 Male: 168, Female: 165


# Identifico indices del df con un valor nulo en la columna sex
indices = df[df["sex"].isnull()].index
# Para cada índice nulo, según ocupe una posicíón par/impar en la lista_indices,
# le asigno un valor 'Female' o 'Male'. De esta forma mantengo la proporción
# de los valores en el conjunto de datos.
#
for n, i in enumerate(indices):
    if (n % 2 == 0):
        df.loc[i, "sex"] = "Female"
    else:
        df.loc[i, "sex"] = "Male"

# Reemplazamos los nulos de bill_length_mm por su promedio
df['bill_length_mm'] = df['bill_length_mm'].fillna(df['bill_length_mm'].mean())
# Reemplazamos los nulos de bill_depth_mm por su promedio
df.fillna({'bill_depth_mm': df['bill_depth_mm'].mean()}, inplace=True)
# Reemplazamos los nulos de flipper_length_mm por su promedio
df['flipper_length_mm'] = \
                df['flipper_length_mm'].fillna(df['flipper_length_mm'].mean())
# Reemplazamos los valores nulos de body_mass_g por su promedio
df.fillna({'body_mass_g':df['body_mass_g'].mean()}, inplace=True)

# Separamos matriz de variables independientes (X) y variable dependiente (y)
y = df["species"]
X = df.drop("species", axis=1)

# Trato las variables independientes categóricas: island y sex y creo las
# variables dummy 'island_Dream', 'island_Torgersen', 'sex_Male'
X = pd.get_dummies(X, drop_first=True)
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.4, \
                                                   random_state=42)
```

#### Ejercicio 5.3
Elige un modelo de clasificación entre los siguientes:
 - Decision tree (Árbol de decisión)
 - Random forest
 - SVM (Support Vector Machine)

Entrenalo con los conjuntos de datos separados en el ejercicio anterior.

```python
# Resuelve aquí el ejericio 5.3
# Eligo Random Forest para este ejercicio por que el algoritmo no requiere \
# tratar las variables categoricas.
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
```

#### Ejercicio 5.4
Utiliza el modelo entrenado en el apartado anterior para predecir los valores de la variable objetivo.

```python
# Resuelve aquí el ejericio 5.4
y_pred = rfc.predict(X_test)
y_pred
```
