# Matplotlib

### Mi primer gr√°fico

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = x **2
```

```python
plt.plot(x, y)
# no es necesario en Jupiter notebook poner plot.show()
```

### Los colores de las l√≠neas

[List of named colors](https://matplotlib.org/stable/gallery/color/named_colors.html)

```python
plt.plot(x, y, 'mediumpurple') # grenn, lightgreen, mediumpurple
```

### El t√≠tulo

```python
plt.plot(x, y, 'olivedrab')
plt.title('Mi primer gr√°fico')
```

### Ejes: Etiquetas

```python
plt.plot(x, y, 'olivedrab')
plt.title('Mi primer gr√°fico')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
```

### Multigr√°ficos en el eje horizontal

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'blue')
plt.title('Mi segundo gr√°fico')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.subplot(1,2,2)
plt.plot(y, x, 'red')
plt.title('Mi tercer gr√°fico')
plt.xlabel('Eje y')
plt.ylabel('Eje x')
```

### Multigr√°ficos en el eje vertical

```python
plt.subplot(2,1,1)
plt.plot(x, y, 'blue')
plt.title('Mi segundo gr√°fico')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.subplot(2,1,2)
plt.plot(y, x, 'red')
plt.title('Mi tercer gr√°fico')
plt.xlabel('Eje y')
plt.ylabel('Eje x')
```

### Multiples gr√°ficos (unos dentro de otros)

```python
figura = plt.figure()
g1 = figura.add_axes([0.1, 0.1, 0.9, 0.9])
g1.plot(x, y, 'blue')
g1.set_title('Mi segundo gr√°fico')
g1.set_xlabel('Eje X')
g1.set_ylabel('Eje Y')
g2 = figura.add_axes([0.2, 0.5, 0.4, 0.4])
g2.plot(y, x, 'red')
g2.set_title('Mi tercer gr√°fico')
g2.set_xlabel('Eje y')
g2.set_ylabel('Eje x')
```

### Dos Series de datos en la misma figura

```python
figura = plt.figure(figsize=(6,6))
graficos = figura.add_axes([0.1, 0.1, 0.9, 0.9])
graficos.plot(x, x**2, 'red', label='Al cuadrado')
graficos.plot(x, x**3, 'green', label='Al cubo')
graficos.plot(x, x**4, 'blue', label='A la cuarta')
graficos.legend(loc = 0)
```

### color hexadecimal del grafico

```python
figura = plt.figure(figsize=(6,6))
graficos = figura.add_axes([0.1, 0.1, 0.9, 0.9])
graficos.plot(x, x**2, color='#ff0000', label='Al cuadrado')
graficos.plot(x, x**3, '#00ff00', label='Al cubo')
graficos.plot(x, x**4, '#0000ff', label='A la cuarta')
graficos.legend(loc = 0)
```

### Grosor, transparencia y tipo de l√≠nea
[Estilos de l√≠nes soportados](https://matplotlib.org/3.0.3/gallery/lines_bars_and_markers/line_styles_reference.html)

```python
figura = plt.figure(figsize=(6,6))
graficos = figura.add_axes([0.1, 0.1, 0.9, 0.9])
graficos.plot(x, x**2, color='#ff0000', linewidth=5, alpha=0.1, linestyle='-', label='Al cuadrado') # '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
graficos.plot(x, x*3, color='#00ff00', linewidth=3, alpha=0.4, linestyle=':', label='Al cubo')
graficos.plot(x, x*4, color='#0000ff', linewidth=10, alpha=0.7, linestyle='-.', label='A la cuarta')
graficos.legend(loc = 0)
```

### Marcadores
[Estilos de marcadores](https://matplotlib.org/3.3.3/api/markers_api.html)

```python
import matplotlib.markers as markers
markers.MarkerStyle.markers
x = np.linspace(0, 10, 20)
y = x **2
figura = plt.figure(figsize=(6,6))
graficos = figura.add_axes([0.1, 0.1, 0.9, 0.9])
graficos.plot(x, y, color='#ff0000', marker=markers.CARETDOWNBASE, markersize=11, linewidth=5, alpha=0.1, linestyle='-', label='Al cuadrado')
graficos.plot(x, x*2, color='#00ff00', marker='s', markersize=11, markerfacecolor='red', markeredgecolor='black',linewidth=3, alpha=1, linestyle=':', label='Al doble')
graficos.legend(loc = 0)
```
