## Grabar un dataframe  en SQL

Hay que instalar sqlalchemy para crear un motro de base de datos en memoria

```python
!pip install sqlalchemy
```

```python
from sqlalchemy import create_engine
```

```python
# Creamos un motor sqlite en memoria.
engine = create_engine('sqlite:///:memory:')
```

```python
# cambio los nombres de las columnas de la tabla html importada para que sean más adecuados a las consultas sql posteriores
db_ready_names={'Población':'Poblacion',
  'Presupuesto (€)':'Presupuesto',
  'Porcentaje población':'Porcentaje_poblacion',
  'Densidad (hab./km²)':'Hab_km2',
  'Superficie (km²)[2]\u200b':'Superficie',
  'Porcentaje superficie':'Porcentaje_superficie',
  'Ciudad más poblada*':'Ciudad_mas_poblada',
  'Comunidad autónoma':'Comunidad_autonoma'}
df.rename(columns=db_ready_names, inplace=True)
df.head()
```

```python
# borro las columnas "Escudo" y "Mapa"
df.drop(['Escudo', 'Mapa'], axis=1, inplace=True)
df.head()
```

```python
# borro la linea 52 que contiene los totales
df.drop(52, axis=0, inplace=True)
df.tail()
```

```python
# inserto los datos en la tabla provincias del motor de base de datos
df.to_sql('provincias', con=engine, index=False)
```

```python
# recupero los datos guardados en la tabla
datos_leidos = pd.read_sql('provincias', con=engine)
datos_leidos
```

```python
# Ejecuto una query sql para obtener el nombre, la poblacion y la superficie de las provincias que empiezan por A
query = 'SELECT Provincia, Poblacion, Superficie FROM provincias WHERE Provincia LIKE "A%"'
datos_leidos = pd.read_sql(query, con=engine)
datos_leidos
```
