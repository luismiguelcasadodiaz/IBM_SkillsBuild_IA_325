## Datos del test

````python
nuevas_muestras = pd.DataFrame({
    'LargoSepalo': [5.1, 6.0, 4.4],
    'AnchoSepalo': [3.5, 2.9, 3.2],
    'LargoPetalo': [1.4, 4.5, 1.3],
    'AnchoPetalo': [0.2, 1.5, 0.2]
})
nuevas_muestras
````

## Definición de la función

````python
def knn_clasificacion(datos, k=3):
  y = datos['species']
  X = datos.drop('species', axis=1)
  KNN = KNeighborsClassifier(n_neighbors=k)
  KNN.fit(X,y)
  return KNN
````

## Preparación del DataFrame con los datos

````python
df_propiedades  = pd.DataFrame(iris.data, columns=iris.feature_names)
df_propiedades.columns = nuevas_muestras.columns
df_target  = pd.DataFrame(iris.target, columns=['species'])
df_target['species'] = df_target['species'].apply(lambda x: iris.target_names[x])
datos= pd.concat([df_propiedades, df_target], axis=1)
datos.head()
````

# Uso del modelo

````python
import numpy as np
KNN = knn_clasificacion(datos, k=3)
predicciones = KNN.predict(nuevas_muestras)
print([str(pred) for pred in predicciones])
````
