## Importaci√≥n de las librer√≠as

````python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
````

## Definici√≥n de la clase SimuladorDatos

````python
class SimuladorDatos:
    """
    Esta clase se encarga de generar un conjunto de datos simulados
    para el problema de predicci√≥n de abandono escolar.

    El conjunto de datos contiene caracter√≠sticas como edad, horas de estudio,
    asistencia, promedio y uso online, junto con la variable objetivo 'Abandono'.
    """
    def __init__(self, n=200, seed=42):
        """
        Inicializa la clase SimuladorDatos.

        Args:
            n (int): Registros (estudiantes) a generar. Por defecto es 200.
            seed (int): La semilla de n√∫meros aleatorios, asegurando la
                        reproducibilidad de los datos. Por defecto es 42.
        """
        self.n = n
        self.seed = seed
        self.df = None
    def generar(self)->pd.DataFrame:
        """
        Genera un conjunto de datos simulados en formato pandas.DataFrame.

        El DataFrame incluye las siguientes columnas con valores aleatorios:
        - Edad: Entero aleatorio entre 18 y 29.
        - Horas_estudio: Decimal aleatorio entre 0 y 30.
        - Asistencia: Decimal aleatorio entre 50 y 100.
        - Promedio: Decimal aleatorio entre 5 y 10.
        - Uso_online: Decimal aleatorio entre 0 y 15.
        - Abandono: Binario (0 o 1). 1 indica abandono (con probabilidad 0.3)
                    y 0 indica que no abandona (con probabilidad 0.7).

        Returns:
            pd.DataFrame: El DataFrame generado con los datos simulados.
        """
        np.random.seed(self.seed)
        self.df = pd.DataFrame()
        self.df['Edad'] = np.random.randint(18, 30, self.n)
        self.df['Horas_estudio'] = np.random.uniform(0, 30, self.n)
        self.df['Asistencia'] = np.random.uniform(50, 100, self.n)
        self.df['Promedio'] = np.random.uniform(5, 10, self.n)
        self.df['Uso_online'] = np.random.uniform(0, 15, self.n)
        self.df['Abandono'] = np.random.choice([0, 1], self.n, p=[0.7, 0.3])
        return self.df
````

````python
sd = SimuladorDatos(5)
df = sd.generar()
df.head()
````

## Definici√≥n de la clase ModeloAbandono

````python
class ModeloAbandono:
    """
    Esta clase gestiona el entrenamiento, evaluaci√≥n y predicci√≥n
    utilizando un modelo de √°rbol de decisi√≥n para predecir el abandono escolar.
    """
    def __init__(self, max_depth=4, random_state=42):
        """
        Inicializa la clase ModeloAbandono.

        Args:
            max_depth (int): La profundidad m√°xima del √°rbol de decisi√≥n.
                             Por defecto es 4.
            random_state (int): La semilla para el generador de n√∫meros aleatorios,
                                asegurando la reproducibilidad del modelo.
                                Por defecto es 42.
        """
        self.max_depth = max_depth
        self.random_state = random_state
        self.modelo = None
        self.X = None
        self.y = None
        self.X_train = None
        self.X_test = None
        self.y_train = None
        self.y_test = None
        self.y_pred = None

    def entrenar(self, data:pd.DataFrame):
        """
        Entrena el modelo de √°rbol de decisi√≥n.

        Divide los datos en conjuntos de entrenamiento y prueba,
        y ajusta el modelo a los datos de entrenamiento.

        Args:
            data (pd.DataFrame): DataFrame que contiene los datos de estudiantes,
                                 incluyendo las caracter√≠sticas y la variable 'Abandono'.
        """
        self.y = data['Abandono']
        self.X = data.drop('Abandono', axis=1)

        # Dividir los datos en conjuntos de entrenamiento y prueba
        print(self.X.head())
        print(self.y.head())
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(self.X, self.y, test_size=0.2, random_state=self.random_state)


        # Inicializar y entrenar el modelo de √°rbol de decisi√≥n
        self.modelo = DecisionTreeClassifier(max_depth=self.max_depth, random_state=self.random_state)
        self.modelo.fit(self.X_train, self.y_train)
        print('Entrenamiento completado.\n')

    def evaluar(self):
        """
        Eval√∫a el rendimiento del modelo entrenado en el conjunto de prueba.

        Calcula y muestra la precisi√≥n del modelo y un reporte de clasificaci√≥n
        que incluye precisi√≥n, recall y f1-score.
        """

        if self.modelo is None:
            print("Error: El modelo no ha sido entrenado. Ejecute el m√©todo entrenar() primero.")
            return

        # Realizar predicciones en el conjunto de prueba
        self.y_pred = self.modelo.predict(self.X_test)

        # Calcular y mostrar la precisi√≥n
        accuracy = accuracy_score(self.y_test, self.y_pred)
        print(f'Precisi√≥n del modelo: {accuracy}')

        # Mostrar el reporte de clasificaci√≥n
        print('Reporte de clasificaci√≥n:')
        print(classification_report(self.y_test, self.y_pred))


    def predecir_estudiante(self, estudiante_df:pd.DataFrame):
        """
        Predice si un estudiante abandonar√° o no sus estudios.

        Args:
            estudiante_df (pd.DataFrame): DataFrame que contiene los datos
                                         de un solo estudiante con las mismas
                                         columnas de caracter√≠sticas que los datos de entrenamiento.

        Returns:
            str: Una cadena que indica si el estudiante "Abandonar√°" o "Seguir√° estudiando".
        """
        if self.modelo is None:
            return "Error: El modelo no ha sido entrenado. Ejecute el m√©todo entrenar() primero."

        # Realizar la predicci√≥n
        prediccion = self.modelo.predict(estudiante_df)
        if prediccion[0] == 0:
            return 'Seguir√° estudiando'
        else:
            return 'Abandonar√°'
````

##Definici√≥n de la clase TestBasicoModeloAbandono

````python
class TestBasicoModeloAbandono:
    def ejecutar(self):
        sd = SimuladorDatos(200, 42)
        data = sd.generar()
        print("Datos simulados:")
        print(data.head())
        ma = ModeloAbandono(4, 42)
        ma.entrenar(data)
        ma.evaluar()
        sd2 = SimuladorDatos(1, 24)
        nuevo_estudiante = sd2.generar()
        nuevo_estudiante_caracterisitcas = nuevo_estudiante.drop('Abandono', axis=1)
        resultado = ma.predecir_estudiante(nuevo_estudiante_caracterisitcas)
        print(f'El estudiante probablemente: {resultado}')
````
