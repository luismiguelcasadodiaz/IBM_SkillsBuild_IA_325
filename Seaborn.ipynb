This plots the residuals of a linear regression. The central plot shows the **difference** between the **actual** values and the values **predicted** by the regression line. The marginal plots show the distribution of these residuals. This is helpful for assessing the goodness of fit of a linear model.

#### Promedio de factura por sexo

This is because sns.barplot by default shows the mean (average) of the y variable for each category in the x variable, along with a confidence interval (the error bars).

```python
import matplotlib.pyplot as plt
sns.barplot(x = 'sex', y = 'total_bill', data = propinas)
plt.title('Promedio de factura por sexo')
```

Here are the different contexts available in sns.set_context():

'paper' (default): This is the default context and is generally suitable for plots that will be included in academic papers or reports. It uses relatively small and concise styling.

'notebook': This context is slightly larger than 'paper' and is well-suited for interactive work in Jupyter notebooks or similar environments where plots are often viewed at a moderate size.

'talk': This context increases the size of labels, lines, and markers, making the plots more readable when presented in a talk or slideshow.

'poster': This is the largest context, designed for plots that will be displayed as posters. It uses very large and prominent elements for maximum visibility from a distance

```python
sns.set_context('poster')
plt.figure(figsize=(10,5))
sns.set_style('ticks') # one of {darkgrid, whitegrid, dark, white, ticks
sns.countplot(x='sex', data=propinas)
sns.despine(left=True, bottom=True)
```

```python
sns.set_context()
sns.set_context( font_scale=2)
plt.figure(figsize=(8,5))
sns.set_style('ticks') # one of {darkgrid, whitegrid, dark, white, ticks
sns.countplot(x='sex', data=propinas)
sns.despine(left=True, bottom=True)
```
